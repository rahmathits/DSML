```python
# Detecting outliers using boxplots for continuous variables
continuous_columns = ['User_ID', 'Occupation', 'Marital_Status', 'Product_Category', 'Purchase']

# Plotting boxplots to visualize outliers
plt.figure(figsize=(15, 10))
for i, column in enumerate(continuous_columns, 1):
    plt.subplot(2, 3, i)
    plt.boxplot(df[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# Assuming df is your DataFrame
continuous_columns = ['User_ID', 'Occupation', 'Marital_Status', 'Product_Category', 'Purchase']

# Calculate the 5th and 95th percentiles and clip the data
for column in continuous_columns:
    lower_bound = np.percentile(df[column], 5)
    upper_bound = np.percentile(df[column], 95)
    df[column] = np.clip(df[column], lower_bound, upper_bound)

print("Data clipped between 5th and 95th percentiles for continuous variables.")

# Verify the clipping
for column in continuous_columns:
    print(f"\n{column} after clipping:")
    print(df[column].describe())
```

```python
from scipy.stats import ttest_1samp

yield_data = [14.3,12.6,13.7,10.9,13.7,12.0,11.4,12.0,12.6,13.1]
alpha = 0.05  #5% significance level

tstat, pvalue = ttest_1samp(yield_data, 12.0, alternative = "greater")
print(pvalue)

if pvalue < alpha:
  print('Reject H0 ; Yield will be more than 12 quintals per hectare')
else:
  print('Fail to reject H0 ; Yield will be 12 quintals per hectare')
```

```python
import pandas as pd
female_scores=pd.Series([25,30,45,49,47,35,32,42])

male_scores=pd.Series([45,47,25,22,29,32,27,28,40,49,50,33])
from scipy.stats import ttest_ind
alpha = 0.02  #2% significance level

tstat, pvalue = ttest_ind(female_scores, male_scores, alternative = "greater")
print(pvalue)

if pvalue < alpha:
  print('Reject H0 ; Average is same for male and female')
else:
  print('Fail to reject H0 ; Average of females is greateer than avg score of Males ')
```

```python
Ammonium_chloride = [13.4, 10.9, 11.2, 11.8, 14, 15.3, 14.2, 12.6, 17, 16.2, 16.5, 15.7]
Urea = [12, 11.7, 10.7, 11.2, 14.8, 14.4, 13.9, 13.7, 16.9, 16, 15.6, 16]
```

```python
from scipy.stats import ttest_ind
alpha = 0.02  #2% significance level

tstat, pvalue = ttest_ind(Ammonium_chloride, Urea, alternative = "two-sided")
print(pvalue)

if pvalue < alpha:
  print('Reject H0 ; The effect of ammonium chloride and urea on grain yield is not equal')
else:
  print('Fail to reject H0 ; The effect of ammonium chloride and urea on grain yield is equal ')
```

```python
wt_before = [85, 74, 63.5, 69.4, 71.6, 65,90,78]

wt_after = [82, 71, 64, 65.2, 67.8, 64.7,95,77]

from scipy.stats import ttest_rel
alpha = 0.10  #2% significance level

tstat, pvalue = ttest_rel(wt_before, wt_after, alternative = "greater")
print(pvalue)

if pvalue < alpha:
  print('Reject H0 ; Reduced weight')
else:
  print('Fail to reject H0 ; There is no weight reduction')
```

```python
before = [2.45, 0.69, 1.80, 2.80, 0.07, 1.67, 2.93, 0.47, 1.45, 1.34]

after = [7.71, 2.17, 5.65, 8.79, 0.23, 5.23, 9.19, 1.49, 4.56, 4.20]
```

```python
from scipy.stats import ttest_rel
alpha = 0.05  #2% significance level

tstat, pvalue = ttest_rel(before, after, alternative = "less")
print(tstat)
print(pvalue)

if pvalue < alpha:
  print('Reject H0 ; No affect on training')
else:
  print('Fail to reject H0 ; Positive affect on training')
```
