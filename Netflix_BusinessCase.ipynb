```python
# Group by country and count unique movie titles
movies_country_count = nf_data.groupby('country')['title'].nunique().reset_index()
movies_country_count.columns = ['Country', 'Number of Movies']
movies_country_count
```

```python
# Sort by number of movies produced and select top 10 countries
top_10_movies_countries = movies_country_count.sort_values(by='Number of Movies', ascending=False).head(10)
top_10_movies_countries
```

```python
# Filter dataset for TV shows
tv_shows_df = nf_data[nf_data['type'] == 'TV Show']
tv_shows_df.head(10)
```

```python
# Group by country and count unique TV show titles
tv_shows_country_count = tv_shows_df.groupby('country')['title'].nunique().reset_index()
tv_shows_country_count.columns = ['Country', 'Number of TV Shows']
tv_shows_country_count
```

```python
# Sort by number of TV shows produced and select top 10 countries
top_10_tv_shows_countries = tv_shows_country_count.sort_values(by='Number of TV Shows', ascending=False).head(10)
top_10_tv_shows_countries
```

```python
# Convert 'date_added' column to datetime format
import pandas as pd
import matplotlib.pyplot as plt
nf_data['date_added'] = pd.to_datetime(nf_data['date_added'], format='%B %d, %Y')
nf_data
```

```python
cast_df = nf_data[["title", "cast"]]
cast_df["unnested_cast"] = cast_df ["cast"].apply(lambda x: str(x).split(", "))
cast_df = cast_df.explode("unnested_cast")
cast_df.head(10)
```

```python
df_exploded = nf_data.explode('cast')
df_exploded
```

```python
# a. Identify the top 10 actors who have appeared in most movies or TV shows
top_actors = df_exploded.groupby('cast')['title'].nunique().reset_index()
top_actors.columns = ['Actor', 'Number of Titles']
top_10_actors = top_actors.sort_values(by='Number of Titles', ascending=False).head(10)
top_10_actors
```

```python
# Identify the top 10 directors who have directed most movies or TV shows
top_directors = nf_data.groupby('director')['title'].nunique().reset_index()
top_directors.columns = ['Director', 'Number of Titles']
top_10_directors = top_directors.sort_values(by='Number of Titles', ascending=False).head(10)
top_10_directors
```

```python
# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.barh(top_10_actors['Actor'], top_10_actors['Number of Titles'], color='skyblue')
plt.xlabel('Number of Titles')
plt.ylabel('Actor')
plt.title('Top 10 Actors in Most Movies/TV Shows')

plt.subplot(1, 2, 2)
plt.barh(top_10_directors['Director'], top_10_directors['Number of Titles'], color='lightgreen')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.title('Top 10 Directors in Most Movies/TV Shows')

plt.tight_layout()
plt.show()
```

```python
nf_data['date_added'] = pd.to_datetime(nf_data['date_added'],format='mixed')
nf_data
```

```python
# Calculate the difference in days between 'date_added' and 'release_year'
nf_data['days_to_add'] = (nf_data['date_added'] - pd.to_datetime(nf_data['release_year'], format='%Y')).dt.days
nf_data
```

```python
# Filter out missing values and negative differences (if any)
df_valid = nf_data[nf_data['days_to_add'].notna() & (nf_data['days_to_add'] >= 0)]
df_valid
```

```python
# Calculate the average of 'days_to_add'
avg_days_to_add = df_valid['days_to_add'].mode()[0]
avg_days_to_add
```
