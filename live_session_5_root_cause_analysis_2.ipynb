# Building Clarity in defining Null Hypothesis and Alternate Hypothesis

A software update claims to reduce the average task processing time from 5 seconds.

### Defining Null and Alternative Hypotheses

1. **Null Hypothesis (\(H_0\))**: 
   - This is a statement of no effect or no difference. It represents the status quo or the existing belief.
   - It is what you are trying to test against.
   - It often contains the equal sign = .
   - Example: The average amount of coffee made is 200 ml (\(H_0: \mu = 200\)).

2. **Alternative Hypothesis (\(H_1\) or \(H_a\))**: 
   - This is what you want to prove.
   - It represents a change or difference.
   - It does not contain the equal sign, what it does contain is (<, >, ≠).
   - Example: The average amount of coffee made is less than 200 ml (\(H_a: \mu < 200\)).

### Common Misconceptions and Clarifications

1. **The Claim is Always in the Alternative Hypothesis**:
   - This is a common misconception. The claim can be in the null hypothesis or the alternative hypothesis, depending on the context of the problem.
   - Example 1: A coffee maker claims to make 200 ml of coffee. A competitor claims it makes less. In this case, the claim (making less than 200 ml) is in the alternative hypothesis.
   - Example 2: A fitness app claims that users walk an average of 8,000 steps per day. To test this claim, we would place it in the null hypothesis (\(H_0: \mu = 8000\)) and test for evidence against it.

2. **Why the Equal Sign is in the Null Hypothesis**:
   - The equal sign is in the null hypothesis because hypothesis testing is about challenging the status quo. You start with the assumption that the null hypothesis is true and then use sample data to gather evidence against it.
   - If there is enough evidence (a low enough p-value), you reject the null hypothesis in favor of the alternative hypothesis.

3. **Interpreting P-Values and Making Decisions**:
   - A low p-value (typically less than 0.05) indicates that the sample data provides enough evidence to reject the null hypothesis.
   - A high p-value indicates that there isn’t enough evidence to reject the null hypothesis, and you fail to reject it. This doesn’t mean you accept the null hypothesis as true; it just means there isn’t enough evidence against it.

### Tips

1. **Identify the Claim and the Status Quo**: Determine what is being claimed and what is the existing belief or standard. This will help in correctly setting up the null and alternative hypotheses.
2. **Practice with Examples**: Go through multiple examples with different scenarios to understand how to set up hypotheses in different contexts.
3. **Remember the Role of the Equal Sign**: The null hypothesis will contain the equal sign, while the alternative hypothesis reflects the change or difference you are testing for.

# SQL Interview Question

**Scenario:**
You are working with a database for an online bookstore. The database contains the following tables:

1. **Books**
   - `book_id` (Primary Key)
   - `title`
   - `author_id`
   - `genre`
   - `price`
   - `publish_date`

2. **Authors**
   - `author_id` (Primary Key)
   - `name`
   - `country`

3. **Sales**
   - `sale_id` (Primary Key)
   - `book_id`
   - `customer_id`
   - `sale_date`
   - `quantity`
   - `total_price`
   <!-- is the same book-id sold multiple time, or for multiple books, there will be multiple rows, where sale-id can be same, but book-id -->

4. **Customers**
   - `customer_id` (Primary Key)
   - `name`
   - `email`
   - `join_date`

**Question 1:**
Write a SQL query to generate a report that shows the top 3 genres that had the highest total sales revenue in the last quarter. The report should also include the total number of books sold, the average book price, and the total revenue for each of these genres. Assume the current date is 2023-12-02.

**Question 2:**
Write a SQL query to identify the top 5 authors whose books have generated the highest average revenue per book sold since the beginning of the current year (2023). The query should return the author's name, the average revenue per book, and the total number of books sold by each of these authors. Additionally, provide this information only for authors who have sold more than 50 books in total within this period.



### Answer 1:


```sql
SELECT 
    B.genre,
    COUNT(S.book_id) AS total_books_sold,
    AVG(B.price) AS average_book_price,
    SUM(S.total_price) AS total_revenue
FROM 
    Sales S
JOIN 
    Books B ON S.book_id = B.book_id
WHERE 
    S.sale_date BETWEEN '2023-09-01' AND '2023-11-30'
GROUP BY 
    B.genre
ORDER BY 
    total_revenue DESC
LIMIT 3;
```

### Explanation:

1. **Join Operations:** The query joins the `Sales` and `Books` tables to correlate the sales data with the book details.
    
2. **Date Filtering:** It filters the sales data to consider only the sales from the last quarter (assuming the quarters are standard calendar quarters).
    
3. **Grouping:** The query groups the results by the `genre` of the books.
    
4. **Aggregation:** It calculates the total number of books sold (`COUNT(S.book_id)`), the average book price (`AVG(B.price)`), and the total revenue (`SUM(S.total_price)`) for each genre.
    
5. **Ordering and Limiting:** The results are ordered by `total_revenue` in descending order to get the genres with the highest revenue on top. The `LIMIT 3` clause restricts the output to the top 3 genres.
    



### Answer 2:

```sql
SELECT 
    A.name AS author_name,
    ROUND(SUM(S.total_price) / COUNT(S.book_id), 2) AS avg_revenue_per_book,
    COUNT(S.book_id) AS total_books_sold
FROM 
    Sales S
JOIN 
    Books B ON S.book_id = B.book_id
JOIN 
    Authors A ON B.author_id = A.author_id
WHERE 
    S.sale_date >= '2023-01-01'
GROUP BY 
    A.author_id, A.name
HAVING 
    COUNT(S.book_id) > 50
ORDER BY 
    avg_revenue_per_book DESC
LIMIT 5;
```

### Explanation:

1. **Join Operations:** The query joins the `Sales`, `Books`, and `Authors` tables to correlate the sales data with the book and author details.

2. **Date Filtering:** It filters the sales data to include only the sales from the beginning of the current year (2023).

3. **Grouping:** The query groups the results by the `author_id` and `name` from the `Authors` table.

4. **Aggregation:** It calculates the average revenue per book sold (`SUM(S.total_price) / COUNT(S.book_id)`) and the total number of books sold (`COUNT(S.book_id)`) for each author.

5. **Having Clause:** The `HAVING` clause filters out the authors who have sold 50 or fewer books, focusing on those who have sold more.

6. **Ordering and Limiting:** The results are ordered by `avg_revenue_per_book` in descending order to find authors with the highest average revenue per book. The `LIMIT 5` clause restricts the output to the top 5 authors.

````python
df['RequestHour'] = df['Request timestamp'].dt.hour

df["TimeSlot"] = df["RequestHour"].apply(lambda x: "Dawn" if x<=4 else ("Early Morning" 
                                                                        if x<=9 else ("Noon" 
                                                                                      if x<=16 else ("Late Evening" 
                                                                                                     if x<=21 else "Night"))))
df['Cab Availability'] = df['Status'].apply(lambda x: "Available" if x=="Trip Completed" else "Not Available")

# let's create day of week column
df['RequestDayOfWeekNum'] = df['Request timestamp'].dt.dayofweek

df.head()
````
